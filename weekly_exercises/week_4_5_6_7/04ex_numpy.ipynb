3\. **Matrix masking**

Create a 10 by 6 matrix of float random numbers, distributed between 0 and 3 according to a flat distribution.

After creating the matrix, set all entries $< 0.3$ to zero using a mask.

````python
dim1, dim2 = 10, 6
print("Array:")
flat_arr = ((np.random.rand(dim1,dim2))*3).round(5)
print(flat_arr)
mask = flat_arr < 0.3 
for x in range(dim1):
    for y in range(dim2):
        if mask[x,y]:
            flat_arr[x,y]=0

print("\nFiltered:\n", flat_arr)
````

4\. **Trigonometric functions**

Use `np.linspace` to create an array of 100 numbers between $0$ and $2\pi$ (inclusive).

  * Extract every 10th element using the slice notation
  * Reverse the array using the slice notation
  * Extract elements where the absolute difference between the `sin` and `cos` functions evaluated for that element is $< 0.1$
  * **Optional**: make a plot showing the sin and cos functions and indicate where they are close

````python
trigArr = np.linspace(0,2*math.pi,100)

print("Array: \n",trigArr)

#1 slice
sliceArray=[]
for i in range(9, trigArr.size, 10):
    sliceArray.append(trigArr[i])
print("\n>> 1) Slice:  \n", sliceArray)

#2 reverse
reverse = trigArr[::-1]
print("\n>> 2) Reverse:\n", reverse)

#3 difference
print("\n>> 3) List elements where the absolute difference between the sin and cos functions evaluated for that element is < 0.1")
print("#\tElement value\t\t\tAbsolute difference\t\tArray position")

c = 0

for i in range(trigArr.size):
    difference = abs(math.sin(trigArr[i]) - math.cos(trigArr[i]))
    if ( difference < 0.1):
        c+=1
        print(c, "\t", trigArr[i], "\t\t", difference, "\t\t", i)

#4 plot
print("\n>> 4) Plot (with intersection):")
cos_func, sin_func = np.cos(trigArr), np.sin(trigArr)

fig, ax = plt.subplots()
ax.plot(trigArr, cos_func, 'b-')
ax.plot(trigArr, sin_func, 'r-')

intersections = []
prev_dif = 0
t0, prev_c1, prev_c2 = None, None, None
for t1, c1, c2 in zip(trigArr, cos_func, sin_func):
    new_dif = c2 - c1
    if new_dif * prev_dif < 0: #change sing => there is an intersection
        denom = prev_dif - new_dif
        intersections.append(((-new_dif * t0 + prev_dif * t1) / denom,
                              (c1 * prev_c2 - c2 * prev_c1) / denom))
    t0, prev_c1, prev_c2, prev_dif = t1, c1, c2, new_dif

ax.plot(*zip(*intersections), 'go', alpha=0.7, ms=10)
plt.show()
````

6\. **Broadcasting**

Use broadcasting to create a grid of distances.

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles.

The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Build a 2D grid of distances among each city along Route 66
  * Convert the distances in km

````python
cities = [
    "Chicago ", "Springfield ", "Saint-Louis ", "Tulsa ", "Oklahoma City ",
    "Amarillo ", "Santa Fe ", "Albuquerque ", "Flagstaff ", "Los Angeles"
]
distance_miles = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])
distance_km = np.array([x * 1.609 for x in distance_miles])

#Build a 2D grid of distances among each city along Route 66
miles_op = abs(distance_miles[:, np.newaxis] - distance_miles)
km_op = abs(distance_km[:, np.newaxis] - distance_km)

print(">> Grid in miles")
print(tabulate(miles_op, headers=cities, tablefmt="github", colalign="left"))

print("\n>> Grid in km")
print(tabulate(km_op, headers=cities, tablefmt="github", colalign="left"))
````
